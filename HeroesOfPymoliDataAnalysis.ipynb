Purchasing Analysis (Gender)

```python
# The below each broken by gender
# Purchase Count
# Average Purchase Price
# Total Purchase Value
# Normalized Totals (normalizing for the # of people in each gender group)

#grouped on gender
gender_grouped_df = hop_df.groupby("Gender")

#values to grab each of the items
purchase_count = gender_grouped_df["Price"].count()
purchase_average = round(gender_grouped_df["Price"].mean(),2)
purchase_sum = round(gender_grouped_df["Price"].sum(),2)
normalized_average = round(purchase_sum/purchase_count,2)


# # Find the percentage of bugs overall fixed by each Assignee
# normalized_sum = gender_grouped_df["Gender"].sum()
# normalized_count = gender_grouped_df["Gender"].count()
# normalized_value = round(normalized_sum/normalized_count,2)


# Creating a new DataFrame using both duration and count
purchasing_summary_df = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Average Purchase Price": purchase_average, 
    "Total Purchase Value":purchase_sum,
    "Normalized Totals": normalized_average

})

#organize the dataframe to be in a logical order
purchasing_summary_df_reorganized = purchasing_summary_df[["Purchase Count", 
                                                             "Average Purchase Price", 
                                                             "Total Purchase Value", 
                                                            "Normalized Totals"]]
purchasing_summary_df_reorganized
```

Age Demographics

```python
# The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
# Purchase Count
# Average Purchase Price
# Total Purchase Value
# Normalized Totals (normalizing for the # of people in each age group)

#figure out range of bin
hop_df["Age"].max()
hop_df["Age"].min()

# Create the bins in which Data will be held
bins = [0,10, 15, 20, 25, 30, 35, 40, 45, 100]
bin_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40-44","40+"]
hop_df["Age Bin"] = pd.cut(hop_df["Age"],bins, labels=bin_labels)


#grouped on bin
age_grouped_df = hop_df.groupby("Age Bin")

#values to grab each of the items
age_purchase_count = age_grouped_df["Price"].count()
age_purchase_average = round(age_grouped_df["Price"].mean(),2)
age_purchase_sum = round(age_grouped_df["Price"].sum(),2)
age_normalized_average = round(age_purchase_sum/age_purchase_count,2)

# Creating a new DataFrame using both duration and count
age_purchasing_summary_df = pd.DataFrame({
    "Purchase Count": age_purchase_count,
    "Average Purchase Price": age_purchase_average, 
    "Total Purchase Value":age_purchase_sum,
    "Normalized Totals": age_normalized_average
})

#organize the dataframe to be in a logical order
age_purchasing_summary_df_reorganized = age_purchasing_summary_df[["Purchase Count", 
                                                             "Average Purchase Price", 
                                                             "Total Purchase Value", 
                                                            "Normalized Totals"]]
age_purchasing_summary_df_reorganized
```

Most Popular Items

```python
# Identify the 5 most popular items by purchase count, then list (in a table):
# Item ID
# Item Name
# Purchase Count
# Item Price
# Total Purchase Value


#top items with id
top_items = hop_df.groupby("Item ID")["Item Name"].count().nlargest(5).reset_index()
item_total = hop_df.groupby("Item ID")["Price"].sum().reset_index()
item_price = hop_df.groupby("Item ID")["Price"].mean().reset_index()
item_lookup = hop_df[["Item ID","Item Name"]]


# Merge three dataframes using a left join
merge_table = pd.merge(top_items, item_total, on= "Item ID")
merge_table_two = pd.merge(merge_table, item_price, on= "Item ID")

indexed = merge_table_two.set_index("Item ID")



#rename columns
indexed_renamed = indexed.rename(columns={'Price_y': 'Item Price',
                                                        'Price_x': 'Total Purchase Value',
                                           'Item Name': "Purchase Count"})
# #reorganize columns
# indexed_renamed_reorganized = indexed_renamed[["Item Name",
#                                                "Purchase Count",
#                                                "Item Price", 
#                                                 "Total Purchase Value"]]
# indexed_renamed_reorganized
indexed_renamed
```

Most Profitable Items


```python
# Identify the 5 most profitable items by total purchase value, then list (in a table):
# Item ID
# Item Name
# Purchase Count
# Item Price
# Total Purchase Value


#top items with id
top_items_spend = hop_df.groupby("Item ID")["Price"].sum().nlargest(5).reset_index()
top_items_count = hop_df.groupby("Item ID")["Item Name"].count().reset_index()
top_item_lookup = hop_df[["Item ID","Item Name"]]


# Merge three dataframes using a left join
merge_table_profit = pd.merge(top_items_spend, top_items_count, on= "Item ID")
merge_table_two_profit = pd.merge(merge_table_profit, top_item_lookup, on= "Item ID")
merge_table_two_profit

indexed_profit = merge_table_two_profit.set_index("Item ID")
indexed_profit


#rename columns
indexed_renamed_profit = indexed_profit.rename(columns={'Item Name_x': 'Purchase Count',
                                                        'Price': 'Total Purchase Value',
                                          'Item Name_y': "Item Name"
                                                        
                                                        })
#reorganize columns

indexed_renamed_profit_reorganized = indexed_renamed_profit[["Item Name",
                                               "Purchase Count",
                                                "Total Purchase Value"]]
indexed_renamed_profit_reorganized
```
